```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
top_k_levels = ["top-5", "top-10", "top-15", "top-20", "top-25"]
low_affinity = {}
low_affinity["protein"] = [8.0, 9.0, 11.7, 12.3, 11.0]
low_affinity["ligand"] = [41.0, 49.5, 53.3, 53.8, 54.4]
low_affinity["interaction"] = [51.0, 41.5, 35.0, 33.8, 33.2]

medium_affinity = {}
medium_affinity["protein"] = [10.0, 11.0, 12.0, 11.5, 12.0]
medium_affinity["ligand"] = [33.0, 36.0, 38.0, 41.5, 42.6]
medium_affinity["interaction"] = [57.0, 53.0, 50.0, 47.0, 45.4]

high_affinity = {}
high_affinity["protein"] = [12.0, 12.0, 12.7, 11.5, 11.8]
high_affinity["ligand"] = [37.0, 37.5, 36.3, 36.0, 36.6]
high_affinity["interaction"] = [51.0, 50.5, 51.0, 52.5, 51.6]
```

```python
affinity_levels = ["Low-affinity", "Medium-affinity", "High-affinity"]

for affinity in affinity_levels:

    affinty_level_dict = low_affinity if affinity == "Low-affinity" else medium_affinity if affinity == "Medium-affinity" else high_affinity


    fig = figure(figsize=(20, 10))

    ax = plt.subplot(111)

    sns.set(font_scale=2.5)

    sns.lineplot(x=top_k_levels, y=affinty_level_dict["protein"], label="Protein", color="darkblue", marker="o", linewidth=2.5)
    sns.lineplot(x=top_k_levels, y=affinty_level_dict["ligand"], label="Ligand", color="darkred", marker="o", linewidth=2.5) #with font 2.5 linewidth=3 and figsize = (20,10)
    sns.lineplot(x=top_k_levels, y=affinty_level_dict["interaction"], label="Interaction", color="darkgreen", marker="o", linewidth=2.5)

    y_max = max(max(affinty_level_dict["protein"]), max(affinty_level_dict["ligand"]), max(affinty_level_dict["interaction"]))
    plt.ylim(0, y_max + 12) #with font 2.5 y_max + 13, with font 2.0 y_max + 11
    plt.xlabel("Top-k edges")
    plt.ylabel("Edges (%)")
    plt.title(affinity +  " interactions")
    # plt.legend(loc="upper right")
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1),
          ncol=3, fancybox=True, shadow=True)

    plt.savefig("results/plots" + "/"+ affinity + "_top_k_increased_fonts_2.5_.png", dpi=300)
    plt.show()
    plt.close()
```
